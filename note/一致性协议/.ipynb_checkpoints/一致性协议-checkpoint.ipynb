## 1.  2PC与3PC
分布式系统中节点的角色：
  * 协调者：统一调度分布式节点的执行逻辑
  * 参与者：被调度的节点。
  
### 1.1 2PC两阶段提交

 绝大部分的关系型数据库采用两阶段提交的方式，来完成分布式事务处理。协议说明如下
#### 阶段一：提交事务请求
 1. 事务询问 ：协调者向所有参与者发送事务请求内容，询问是否可以进行实务操作
 2. 执行事务 ： 参与者执行请求事务，并且记录事务日志（Undo与Redo信息）。
 3. 各参与者向协调者反馈事务询问的响应：若参与者状态正常 ，返回yes可执行，否则返回no不可执行。
 
#### 阶段二：执行事务提交，分以下几种情况：

 1. 执行事务提交：若所有的参与者都返回yes，则执行事务提交。
     + 发送事务请求：协调者向所有的参与者发出commit请求。
     + 事务提交 参与者接收到commit请求后，会正式执行事务提交操作，并在完成提交后释放在整个事务提交期间占用的事务资源。
     + 反馈结果：向协调者发送Ack，确定事务执行完成
     + 完成事务：协调者接收到所有Ack后，该事务执行完成。
     
 
 2. 中断事务：任何一个参与者返回No响应，或者等待超时，无法接收到所有参与者的反馈，就会中断事务：
     + 发送回滚请求：向所有节点发送出rollBack请求。
     + 参与者收到回滚请求，会进行事务回滚
     + 参与者回滚后，发出Ack请求
     + 中断事务：协调者收到所有的Ack响应后，完成事务中断
     
####   优缺点
 * 优点：原理简单，易于实现
 * 缺点：同步阻塞，单点问题，网络分裂，过于保守。
 
 
### 3PC
    
#### 阶段一：canCommit
* 事务询问
* 各参与者反馈事务询问的响应
#### 阶段二： PreCommit
* 执行事务预提交：所有参与者都反馈Yes，则执行预提交
 1. 发送预提交请求
 2. 食物与提交
 3. 各个参与者反馈事务执行的响应。
 * 中断事务
