#### 分布式系统
##### 分布式系统的特点
* 定义：分布式系统是一组硬件或软件分布在不同的网络计算机上，彼此之间仅仅通过消息传递进行通信和协调的系统
* 特性：
    * 分布性：分布式系统中的计算机在空间上随机分布
    * 对等性：分布式系统中的计算机节点没有主从之分，所有计算机节点都是对等的。
    * 并发性：同一个分布式系统中的计算机，可能会操作一些共享的资源，如数据库。
    * 缺乏全局时钟：难定以事件发生的先后
    * 故障总会发生：
##### 分布式环境的各种问题
* 通信异常：节点之间进行网络通信，有网络异常的风险。
* 网络分区：网络发生异常时，可能造成分布式系统中一部分节点之间可以通信正常，另一部分无法连接的情况。会导致系统内部出现小集群。
* 三态：分布式系统的每一次请求和响应存在的三种状态：**成功，失败，超时**。
    超时的原因：
        1. 网络原因，导致请求消息丢失
        2. 服务端收到消息，返回消息时，由于网络原因，导致响应消息丢失
* 节点故障：单个节点出现的宕机或僵死现象



##  2 ACID与CAP/BASE
###  2.1ACID
 * 事务：一系列对系统中数据进行访问和更新的操作组成的一个程序执行逻辑单元。
     有如下用处：
     1. 多个程序并发访问数据时，事务可以为这些应用程序提供一个隔离方法
     2. 提供了从异常恢复到正常状态的方法
* ACID的含义
 1. 原子性
 2. 一致性：事务的执行不能破坏数据库的完整和一致性状态，即事务的执行，必须使数据库从一个一致性状态转变到另一个一致性状态。
 3. 隔离性：并发的环境中，并发的事务是相互隔离的，即一个数据库事务内部的操作以及使用的数据对其他事务是隔离的，并发执行时，两个事务不互相干扰。
     SQL定义了如下四个隔离级别：
     串行化：最高级别，所有事务不能并发执行。
     可重复读：允许不可重复读。保证一个事务在处理过程中，多次读取一个值，读到的值始终一致。即保证在事务处理中，其他事务不可对其使用到的值进行更新操作（类似读写锁，只针对update，不针对insert）
     读已提交（授权读）：允许幻读，不可重复读。只允许读取一个已经被提交的事务（update操作更新存在的一行数据）。
     读未提交（未授权读）：允许脏读，不可重复读，幻读，即一个事物，允许读取尚未完成的另一个事务更新未提交的值（update操作更新存在的一行数据）。
     
 4. 持久性：事务一旦提交，其所做的对应状态的更改就是永久性的（写入物理存储）。
 
### 2.2分布式事务

 * 是指事务的参与者，支持事务的服务器，资源服务器，以及事务管理器分别位于分布式系统的不同节点上。通常一个分布式事务会涉及多个数据源或业务系统的操作。
* 分布式事务可以看作是多个分布式操作序列（子事务）组成。实现分布式事务的ACID更加复杂。

###  2.3CAP和BASE理论
——分布式系统的ACID实现：会产生系统可用性和严格一致性的冲突。关于分布式系统的关于一致性与可用性的理论有CAP，BASE等：
#### CAP：

* 内容：一个分布式系统不可能同时满足一致性（C:consistency），可用性(A；Availability)，和分区容错性（P：Patition tolerance）这三个基本要求，最多只能满足其中两项

  1. 一致性：分布式环境中，数据在多个副本之间能否保持一致的特性。并且在一致性的的状态下执行数据的更新操作后，系统数据仍然保持一致。
        强一致性：如果在分布式系统中，针对一个数据项进行更新操作后，所有的用户都可以读到最新的值，这样的系统被认为具有严格一致性。
        
  2. 可用性：系统提供的服务一直处于可用的状态，对于用户的每一个请求，都能在有限的时间内返回结果。
        + 有限时间内：对于一个用户请求，系统能够在指定的时间（响应时间内）返回对应的处理结果，若超出这个范围，则系统被认为是不可用的。
        + 返回结果：系统在完成用户的请求后，返回一个正常的响应结果。
        
  3. 分区容错性：分布式系统在遇到任何网络分区故障时，仍然能够保证对外提供满足一致性和可用性的服务。除非是整个网络发生了故障。

* CAP定理应用的场景(不同类型的分布式系统实现)：
 + 放弃C：放弃数据的强一致性，保持数据的最终一致性。引入*时间窗口*的概念：分布式系统在多久能够达到数据一致的状态。
 + 放弃A：一旦系统遇到网络分区或其他故障，客户端需要等待一段时间才可以得到正常响应的服务。
 + 放弃P：放弃分区容错性，若想避免，可以把所有节点配置在同一个分布式节点上。但是这样也放弃了系统的扩展性，同时，一台机器宕机会造成所有服务不可用。

* 分布式系统无法放弃P，所以最终其实是在C和A之间进行权衡。


#### BASE理论：
* BASE:对CAP中一致性和可用性权衡的结果，其来源于大规模互联网系统分布式实践的总结，核心思想是：即使无法做到强一致性，但每个应用可以根据自身的业务特点，采用适当的方式来使系统达到最终一致性。

 + BA（Basically Available）基本可用：分布式系统在出现不可预知的故障时，允许损失部分的可用性。如：
     * 响应时间上的损失
     * 功能上的损失
 + S(Soft state) 弱状态：与硬状态相对于，指允许系统中的数据存在中间状态，并认为该中间状态的存在不会影响系统的整体可用性。即允许系统在不同节点的数据副本数据进行同步的过程中存在延时。
 + Eventually consistent：最终一致性：系统中所有的副本数据，在经过一段时间的同步之后，最终可以达到一个一致的状态。<br/>
                 
    一种特殊的弱一致性：系统保证在没有其他新的更新操作的情况下，数据最终一定能够达到一致的状态，所有客户端对系统的访问都能得到最新的值。窗口时间取决于网络延迟，系统负载，和数据复制方案设计等因素。最终一致性在应用中主要存在以下几种类型：
        1. 因果一致性：进程A在更新完某个数据项后通知了进程B，那么进程B之后对该数据项的访问都可以访问到A更新后的最新值，即不能丢失更新。若C与A无因果关系，则没有这种限制。
        2. 读自己所写：进程A更新了一个数据项之后，自己总是能访问到更新过的最新纸。
        3. 会话一致性：系统保证，能在一个有效的会话中实现"读自己所写"的一致性
        4. 单调读一致性:一个进程从系统中读取出一个数据项的某个值之后，之后该进程的访问都不应返回更旧的值。
        5. 单调写一致性：保证一个进程的写请求按照请求顺序被执行。
                 

* BASE与传统事务的ACID特性相反，主要在于牺牲强一致性，来获取系统的可用性。允许数据在一段时间内是不一致的，但最终达到一致状态。
